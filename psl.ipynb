```python
teams = ['Peshawar Zalmi', 'Quetta Gladiators', 'Islamabad United', 'Karachi Kings', 'Lahore Qalandars' ]
```

```python
psl.columns
```

```python
psl.rename(columns={'win_by_runs': 'Win-Batted First', 'win_by_wickets': 'Win-Bowled First'}, inplace=True)
```

```python
dic = {}
for i in teams:
    ans = psl[psl['winner'] == i]['Win-Batted First'].count()
    print(f'When {i} batted first win {ans} times')
    if i not in dic:
        dic[i] = ans        
```

```python
dic
```

```python
bat_first=pd.Series(dic)
```

```python
bat_first
```

```python
# Horizontal Bar Chart for matches won when batting first with cricket-themed colors
bat_first.plot.barh(title='Match won When Batting First', xlabel='No. of Matches', ylabel='Teams', color=['darkgreen', 'lightgreen', 'navy', 'orange', 'gold'])

# Adding labels on the bars
for a, b in zip(range(len(bat_first)), bat_first.to_numpy()):
    plt.text(b + 0.1, a, str(b), va='center')

plt.show()
```

```python
dic2 = {}
for i in teams:
    ans = psl[psl['winner'] == i]['Win-Bowled First'].count()
    print(f'When {i} bowled first win {ans} times')
    if i not in dic2:
        dic2[i] = ans   
```

```python
dic2
```

```python
bowl_first=pd.Series(dic2)
```

```python
bowl_first
```

```python
bowl_first.values
```

```python
# Horizontal bar chart for matches won when bowling first
bowl_first.plot.barh(title='Matches Won when Bowling First', xlabel='No. of Matches', ylabel='Teams', color='green')

# Adding labels on the bars
for a, b in zip(range(len(bowl_first)), bowl_first.values):
    plt.text(b + 0.2, a, str(b), va='center')  # Display the value at the end of the bar

plt.show()


```

```python
# Batting bowling Comprison
bat_bowl_first=pd.Series({'Batting First':bat_first.sum(), 'Bowling First':bowl_first.sum()})
```

```python
# Bar chart for matches won when batting or bowling first with updated colors
bat_bowl_first.plot.bar(title="Matches Won (Batting vs Bowling First)", 
                         xlabel="Strategy", 
                         ylabel="Number of Matches", 
                         color=['#4CAF50', '#FF9800'],  # Green and orange for a vibrant look
                         rot=0)

# Calculating percentages
total_matches = bat_bowl_first.sum()
percentages = (bat_bowl_first / total_matches) * 100

# Adding percentage labels in the center of the bars
for a, b, p in zip(range(len(bat_bowl_first)), bat_bowl_first.to_numpy(), percentages):
    plt.text(a, b / 2, f'{p:.1f}%', ha='center', va='center', color='white')  # Centering percentages

plt.show()


```

```python
dic = {}
for i in teams:
    ans = psl[(psl['innings1'] == i) | (psl['innings2'] == i)]['result'].count()
    won = psl[psl['winner'] == i]['result'].count()
    percent = (won/ans * 100).round(2)
    batting = psl[psl['winner'] == i]['Win-Batted First'].count()
    per_bat = (batting/won * 100).round(2)
    bowling = psl[psl['winner'] == i]['Win-Bowled First'].count()
    per_bowl = (bowling/won * 100).round(2)
    print(f'{i} played {ans} matches and won {won} matches -------> {percent}%')
    print(f'Out of {won} won matches {batting} won on batting first ----------> {per_bat} ')
    print(f'Out of {won} won matches {bowling} won on bowling first ----------> {per_bowl}\n')
    
    if i not in dic:
        dic[i] = {'Matches': ans, 'Won':won, 'Bat': batting, 'Bowl':bowling, 'Win%':percent, 'Win-Bat%':per_bat,\
                 'Win-Bowl%':per_bowl}
```

```python
dic
```

```python
a=pd.Series(dic)
```

```python
a
```

```python
df=pd.DataFrame(dic).T
```

```python
df
```

```python
# Horizontal Bar Chart
df['Win%'].plot(kind='barh', title='Wins Percentage', color='royalblue', xlabel='Win Percentage', ylabel='Teams')

# Adding data labels to the bars
for index, value in enumerate(df['Win%']):
    plt.text(value + 0.5, index, str(value) + '%', va='center')

plt.show()
```

```python
# Grouped Bar Chart
df[['Win-Bat%', 'Win-Bowl%']].plot(kind='bar', title='Matches Won - Batting First vs Bowling First', rot=15, color=['coral', 'lightblue'])

plt.ylabel('Win Percentage')
plt.xlabel('Teams')
plt.grid(axis='y')
plt.show()
```

```python
for i in teams:
    tot = psl[(psl['home'] == i) | (psl['away'] == i)].shape[0]
    win = psl[psl['winner'] == i].shape[0]
    
    print(f'{i} won {win} matches out of {tot} matches')
```

```python
# head to head match comprison when bowled first
for i in teams:
    for j in teams:
                win = psl[(psl['innings2'] == i) & (psl['winner'] == i) &\
                         (psl['innings1'] == j) ].shape[0]
                playd = psl[(psl['innings2'] == i) &\
                           (psl['innings1'] == j) ].shape[0]
                if playd != 0:
                    print(f'{i} Won {win*100/playd : .2f}% of the time when bowled first against {j} -------> Won = {win} Played = {playd} ')
    print('\n')
```

```python
# head to head match comprison when batting first
for i in teams:
    for j in teams:
                win = psl[(psl['innings1'] == i) & (psl['winner'] == i) &\
                         (psl['innings2'] == j) ].shape[0]
                playd = psl[(psl['innings1'] == i) &\
                           (psl['innings2'] == j) ].shape[0]
                if playd != 0:
                    print(f'{i} Won {win*100/playd : .2f}% of the time when batted first against {j} -------> Won = {win} Played = {playd} ')
    print('\n')
```
